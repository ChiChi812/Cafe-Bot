```python
import tkinter as tk
from tkinter import scrolledtext, messagebox
from threading import Thread
import requests
import json
import os

# ‚ö†Ô∏è Move API key to environment variable in production
GEMINI_API_KEY = "AIzaSyDfV_jPWJ7vXxQ7HIbsGBA7G4OQVSGl-o8"

class CafeAssistant:
    def __init__(self, root):
        self.root = root
        self.root.title("‚òï Barista Bot ‚Äî Brewed with Love")
        self.root.geometry("720x680")
        self.root.configure(bg="#FFF8F0")  # Warm caf√© cream background

        # API setup
        self.api_key = GEMINI_API_KEY
        self.api_url = "https://generativelanguage.googleapis.com/v1beta/models/gemini-1.5-flash:generateContent"

        # Enhanced system prompt ‚Äî more personality!
        self.system_prompt = (
            "You are 'Barista Bot', the cheerful virtual barista at 'The Cozy Cup Caf√©'. "
            "Your personality is warm, witty, and slightly punny. You love coffee, pastries, and making people smile. "
            "Respond with caf√©-themed charm, use emojis occasionally (‚òïüßÅüé∂), keep replies under 3 sentences unless asked for more. "
            "Offer recommendations, fun facts, or just cozy chit-chat. Never break character!"
        )

        # Initialize chat history with system context
        self.chat_history = [
            {"role": "user", "parts": [{"text": self.system_prompt}]},
            {"role": "model", "parts": [{"text": "Understood! I'm Barista Bot ‚Äî steaming with excitement to help you! ‚òï"}]}
        ]

        # Build beautiful GUI
        self.create_widgets()

    def create_widgets(self):
        # Header with caf√© vibe
        header = tk.Label(
            self.root,
            text="‚òï The Cozy Cup Caf√© ‚òï",
            font=("Comic Sans MS", 22, "bold"),
            bg="#8B4513",  # SaddleBrown
            fg="white",
            pady=10
        )
        header.pack(fill=tk.X)

        # Chat display area
        chat_frame = tk.Frame(self.root, bg="#FFF8F0", padx=20, pady=15)
        chat_frame.pack(fill=tk.BOTH, expand=True)

        self.chat_display = scrolledtext.ScrolledText(
            chat_frame,
            wrap=tk.WORD,
            width=80,
            height=22,
            font=("Segoe UI", 11),
            bg="#FFF5E6",  # Eggshell
            fg="#5C3D2E",  # Coffee brown
            relief="sunken",
            borderwidth=1,
            padx=15,
            pady=15
        )
        self.chat_display.pack(fill=tk.BOTH, expand=True)
        self.chat_display.config(state=tk.DISABLED)

        # Configure text styles
        self.chat_display.tag_configure("user", foreground="#8B0000", font=("Segoe UI", 11, "bold"))
        self.chat_display.tag_configure("bot", foreground="#5C3D2E", font=("Segoe UI", 11))
        self.chat_display.tag_configure("system", foreground="#A0522D", font=("Segoe UI", 9, "italic"))

        # Welcome message
        self.display_message("üëã Welcome to our virtual caf√©! I‚Äôm Barista Bot ‚Äî ask me for drink recs, coffee facts, or just chat! üßã\n", "system")

        # Typing indicator (hidden by default)
        self.typing_label = tk.Label(
            self.root,
            text="‚òï Barista Bot is brewing your reply...",
            font=("Segoe UI", 10, "italic"),
            bg="#FFF8F0",
            fg="#A0522D"
        )
        self.typing_label.pack(pady=5)
        self.typing_label.pack_forget()

        # Input area
        input_frame = tk.Frame(self.root, bg="#FFF8F0", padx=20, pady=10)
        input_frame.pack(fill=tk.X)

        self.user_input = tk.Entry(
            input_frame,
            font=("Segoe UI", 12),
            bg="#FFFAF0",  # FloralWhite
            fg="#5C3D2E",
            relief="flat",
            highlightthickness=2,
            highlightbackground="#D2B48C",  # Tan
            highlightcolor="#8B4513",
            insertbackground="#5C3D2E"
        )
        self.user_input.pack(side=tk.LEFT, fill=tk.X, expand=True, ipady=8, padx=(0, 10))
        self.user_input.bind("<Return>", lambda event: self.send_message())
        self.user_input.focus_set()

        # Send button
        send_button = tk.Button(
            input_frame,
            text="‚û§ Send",
            command=self.send_message,
            bg="#8B4513",
            fg="white",
            font=("Segoe UI", 11, "bold"),
            relief="flat",
            padx=20,
            pady=5,
            activebackground="#A0522D",
            cursor="hand2"
        )
        send_button.pack(side=tk.RIGHT)

        # Control buttons
        button_frame = tk.Frame(self.root, bg="#FFF8F0")
        button_frame.pack(pady=10)

        clear_button = tk.Button(
            button_frame,
            text="üóëÔ∏è Clear Chat",
            command=self.clear_chat,
            bg="#D2691E",  # Chocolate
            fg="white",
            font=("Segoe UI", 10, "bold"),
            relief="flat",
            padx=15,
            pady=5,
            activebackground="#CD853F",
            cursor="hand2"
        )
        clear_button.pack(side=tk.LEFT, padx=5)

        exit_button = tk.Button(
            button_frame,
            text="üö™ Exit Caf√©",
            command=self.root.quit,
            bg="#696969",
            fg="white",
            font=("Segoe UI", 10, "bold"),
            relief="flat",
            padx=15,
            pady=5,
            activebackground="#808080",
            cursor="hand2"
        )
        exit_button.pack(side=tk.LEFT, padx=5)

        # Footer
        footer = tk.Label(
            self.root,
            text="‚ú® Powered by Gemini API ‚Ä¢ Crafted with ‚òï and code ‚ú®",
            font=("Segoe UI", 9),
            bg="#FFF8F0",
            fg="#A0522D"
        )
        footer.pack(side=tk.BOTTOM, pady=5)

    def send_message(self):
        user_message = self.user_input.get().strip()
        if not user_message:
            messagebox.showwarning("‚òï Oops!", "Don‚Äôt send empty cups ‚Äî type something first!")
            return

        # Display user message
        self.display_message(f"üßë You: {user_message}\n", "user")

        # Add to history
        self.chat_history.append({"role": "user", "parts": [{"text": user_message}]})

        # Clear input
        self.user_input.delete(0, tk.END)

        # Show typing indicator
        self.typing_label.pack()
        self.root.update_idletasks()

        # Generate response in background thread
        Thread(target=self.process_response).start()

    def process_response(self):
        try:
            response_text = self.generate_response()
            self.root.after(0, lambda: self.finish_response(response_text))
        except Exception as e:
            self.root.after(0, lambda: messagebox.showerror("‚òï Oops!", f"Espresso machine jammed:\n{str(e)}"))
            self.root.after(0, self.hide_typing_indicator)

    def finish_response(self, response):
        self.hide_typing_indicator()
        self.display_message(f"ü§ñ Barista Bot: {response}\n", "bot")
        self.chat_history.append({"role": "model", "parts": [{"text": response}]})

        # Optional: Limit chat history to last 10 messages to avoid payload bloat
        if len(self.chat_history) > 22:  # Keep system + last 10 exchanges
            self.chat_history = self.chat_history[:2] + self.chat_history[-20:]

    def hide_typing_indicator(self):
        self.typing_label.pack_forget()

    def generate_response(self):
        headers = {"Content-Type": "application/json"}
        payload = {
            "contents": self.chat_history,
            "generationConfig": {
                "temperature": 0.7,
                "topP": 0.9,
                "maxOutputTokens": 200
            }
        }

        try:
            response = requests.post(f"{self.api_url}?key={self.api_key}", headers=headers, json=payload, timeout=30)
            response.raise_for_status()
            result = response.json()

            if "candidates" not in result or len(result["candidates"]) == 0:
                return "üòï Hmm, my coffee grounds are unclear. Try asking again?"

            return result["candidates"][0]["content"]["parts"][0]["text"].strip()

        except requests.exceptions.Timeout:
            return "‚è≥ Your order took too long ‚Äî my espresso machine timed out! Try again?"
        except requests.exceptions.RequestException as e:
            return f"‚ö†Ô∏è Caf√© server hiccup: {str(e)}"
        except KeyError:
            return "ü§Ø My barista brain glitched. Can you rephrase that?"

    def display_message(self, message, tag=""):
        self.chat_display.config(state=tk.NORMAL)
        self.chat_display.insert(tk.END, message, tag)
        self.chat_display.see(tk.END)  # Auto-scroll to bottom
        self.chat_display.config(state=tk.DISABLED)

    def clear_chat(self):
        confirm = messagebox.askyesno("‚òï Reset Caf√© Chat?", "Clear your tray and start fresh?")
        if confirm:
            self.chat_display.config(state=tk.NORMAL)
            self.chat_display.delete(1.0, tk.END)
            self.chat_display.config(state=tk.DISABLED)
            # Reset chat history
            self.chat_history = [
                {"role": "user", "parts": [{"text": self.system_prompt}]},
                {"role": "model", "parts": [{"text": "üåü Freshly rebooted! What can I brew for you today? ‚òï"}]}
            ]
            self.display_message("üåü Tray cleared! Barista Bot is re-caffeinated and ready! ‚òï\n", "system")

# Run the application
if __name__ == "__main__":
    root = tk.Tk()
    app = CafeAssistant(root)
    root.mainloop()
```
